```python
#new DF
df_improved = df_filtered

#Row 'date diff' = vigencia
df_improved['inicio_vigencia'] = pd.to_datetime(df_improved['inicio_vigencia'], errors = 'coerce', infer_datetime_format=True)
df_improved['termino_vigencia'] = pd.to_datetime(df_improved['termino_vigencia'], errors = 'coerce', infer_datetime_format=True)
df_improved = df_improved.assign(date_diff=(df_improved['termino_vigencia'] - df_improved['inicio_vigencia']).dt.days)
```

```python
def verifica_contratacao(valor):
    nao_contratado = ['Não contratado', 'NÃ£o possui Cobertura', 'não Contratada', 'NÃ£o Contempla', 'NÃ£o contemplado', 'NÃ£o Contratado', 'Sem carro reserva']
    if valor in nao_contratado:
        return 'Não'
    else:
        return 'Sim'

df_improved['vidros_sim'] = df_improved['coberturas_vidros_retorno'].apply(verifica_contratacao)
df_improved['carro_reserva_sim'] = df_improved['carro_reserva_retorno'].apply(verifica_contratacao)
df_improved['assistencia_24h_sim'] = df_improved['assistencia_24h_retorno'].apply(verifica_contratacao)

def validar_assistencias(row):
    if 'Sim' in row.values:  # If 'Sim' is present in any column
        return 'Sim'
    else:
        return 'Não'

df_improved['assistencias_total'] = df_improved[['vidros_sim', 'carro_reserva_sim', 'assistencia_24h_sim']].apply(validar_assistencias, axis=1)

# Validate
df_improved[['coberturas_vidros_retorno', 'vidros_sim', 'carro_reserva_retorno', 'carro_reserva_sim', 'assistencia_24h_retorno', 'assistencia_24h_sim', 'assistencias_total']].head(10)


```

```python
#new DF
df_improved2 = df_improved

#generate 'faixa_de_edade'
df_improved2['segurado_data_nascimento'] = pd.to_datetime(df_improved2['segurado_data_nascimento'], infer_datetime_format=True)

data_atual = datetime.now()
df_improved2['idade'] = (data_atual - df_improved2['segurado_data_nascimento']).astype('<m8[Y]')

def faixa_idade_range(valor):
    if valor < 17:
        return '1 a 16'
    elif valor < 27:
        return '17 a 26'
    elif valor < 37:
        return '27 a 36'
    elif valor < 47:
        return '37 a 46'
    elif valor < 57:
        return '47 a 56'
    elif valor < 67:
        return '57 a 66'
    elif valor < 77:
        return '67 a 76'
    else:
        return '77 +'
    
df_improved2['faixa_idade'] = df_improved2['idade'].apply(faixa_idade_range)


#generate 'faixa_de_edade_carro'

df_improved2['ano_modelo'] = df_improved2['ano_modelo'].astype(str)
df_improved2['ano_modelo'] = df_improved2['ano_modelo'].astype(int)
df_improved2['idade_carro'] = 2023 - df_improved2['ano_modelo']



def faixa_idade_carro_range(valor):
    if valor < 2:
        return '0 a 1,9 anos'
    elif valor < 4:
        return '2 a 3,9 anos'
    elif valor < 6:
        return '4 a 5,9 anos'
    elif valor < 8:
        return '6 a 7,9 anos'
    else:
        return '8 ou mais anos'

df_improved2['faixa_idade_carro'] = df_improved2['idade_carro'].apply(faixa_idade_carro_range)


#clean 'Female' #error in data set
def clean_segurado_sexo(sexo):
    if sexo == 'female':
        return 'F'
    else:
        return sexo
    
df_improved2['segurado_sexo'] = df_improved2['segurado_sexo'].apply(clean_segurado_sexo)

#generate 'kit_gas'
def verifica_gnv(valor):
    if isinstance(valor, str) and 'gnv' in valor.lower(): 
        return 'Sim'
    else:
        return 'Não'

df_improved2['kit_gas'] = df_improved2['tipo_combustivel'].apply(verifica_gnv)

#generate 'faixa_clase_bonus'
def classe_bonus_range(valor):
    if valor == 0:
        return ' 0 '
    elif valor < 4:
        return '1 a 3'
    elif valor < 8:
        return '4 a 7'
    else:
        return '8 a 10'
    
df_improved2['faixa_clase_bonus'] = df_improved2['bonus_atual'].apply(classe_bonus_range)
 
#generate 'codigo_regiao'
def extrair_codigo_regiao(cep):
    cep_str = str(cep)  # Converter o valor para string
    if len(cep_str) == 8:
        return cep_str[:2]
    else:
        return cep_str[:1]


df_improved2['codigo_regiao'] = df_improved2['cep_circula'].apply(extrair_codigo_regiao)


#generate 'codigo_cidade'
def extrair_cidade(cep):
    cep_str = str(cep)  # Converter o valor para string
    if len(cep_str) == 8:
        return cep_str[:5]
    else:
        return cep_str[:4]


df_improved2['codigo_cidade'] = df_improved2['cep_circula'].apply(extrair_cidade)




#validate
df_improved2[['ano_modelo','idade_carro','faixa_idade_carro','tipo_combustivel','kit_gas','bonus_atual','faixa_clase_bonus', 'cep_circula', 'codigo_regiao','codigo_cidade', 'idade', 'faixa_idade']].head(5)
```

```python
# Creates a new categorical variable 'grupo' based on the unique values of 'id_cotacao'.
df_filtered2['grupo'] = df_filtered2.groupby('id_cotacao').ngroup()

# Creates a new categorical variable 'categoria' based on combinations of 'grupo', 'date_diff', 'tipo_cobertura_retorno', 'assistencias_total', and 'franquia_retorno'.
df_filtered2['categoria'] = df_filtered2.groupby(['date_diff', 'tipo_cobertura_retorno', 'assistencias_total', 'franquia_retorno']).ngroup()

#Creates a new categorical variable 'filtro' based on combinations of 'codigo_fipe', 'codigo_regiao', 'segurado_sexo', 'faixa_idade', 'kit_gas', 'veiculo_blindado', and 'faixa_clase_bonus'.
df_filtered2['filtro'] = df_filtered2.groupby(['codigo_fipe', 'codigo_regiao', 'segurado_sexo', 'faixa_idade','kit_gas', 'veiculo_blindado','faixa_clase_bonus']).ngroup()

# Filters the DataFrame to exclude rows where 'marcas_tipo' is equal to 'moto'.
df_filtered2_moto = df_filtered2[df_filtered2['marcas_tipo'] != 'moto']

# Filters the groups with fewer than 5 rows using the 'filter' method.
df_filtered2_filtrados = df_filtered2_moto.groupby(['grupo', 'categoria', 'filtro']).filter(lambda x: len(x) > 3)

# Groups the filtered DataFrame by 'filtro', 'grupo', and 'categoria' and calculates the median of 'premio_ajustado'.
resultados2 = df_filtered2_filtrados.groupby(['filtro', 'grupo', 'categoria']).agg({'premio_ajustado': ['mean']})

# Merges the results back into the original DataFrame.
df_filtered2_filtrados = df_filtered2_filtrados.merge(resultados2, on=['grupo', 'categoria'])

```

```python
# Calculates the 'desv_da_mediana' for each row, which is the difference between 'premio_ajustado' and the median value for the group, divided by the median value.
df_filtered2_filtrados['desv_da_media'] = (df_filtered2_filtrados['premio_ajustado'] - df_filtered2_filtrados[('premio_ajustado', 'mean')]) / df_filtered2_filtrados[('premio_ajustado', 'mean')]
```

```python
from scipy import stats

# Calculate the probability density function of the 'desv_da_mediana' column
pdf = df_filtered2_filtrados['desv_da_media'].apply(lambda x: stats.norm.pdf(x, df_filtered2_filtrados['desv_da_media'].mean(), df_filtered2_filtrados['desv_da_media'].std()))

# Normalize the probability density function to sum up to one
pdf_norm = pdf / pdf.sum()

# Add the normalized probability density function to the DataFrame
df_filtered2_filtrados['probabilidade'] = pdf_norm
```

```python
# Calculate the sum of probabilities when desv_da_mediana is lower than -0.2
sum_prob = df_filtered2_filtrados.loc[df_filtered2_filtrados['desv_da_media'] < -0.20, 'probabilidade'].sum()

# Print the sum
print("Sum of probabilities when desv_da_media < -0.2: ", sum_prob)
```

```python
# Define a function to check if the difference between two values is less than 0.1
def is_within_tolerance(a, b, tolerance=0.2):
    return abs((a - b)/b) <= tolerance

# Group the dataframe by 'filtro', 'grupo', 'categoria', and then loop through the groups
for group_name, group_df in df_filtered2_filtrados.groupby(['filtro', 'grupo', 'categoria']):
    # Get the unique values of 'premio_ajustado' in the group
    unique_premios = group_df['premio_ajustado'].unique()
    
    # If there is only one unique value, set the 'preco_validado' column to 'nao_validado'
    if len(unique_premios) == 1:
        group_df['preco_validado'] = 'nao_validado'
    else:
        # Otherwise, loop through the rows in the group and compare 'premio_ajustado' to the other values
        for idx, row in group_df.iterrows():
            other_premios = [p for p in unique_premios if p != row['premio_ajustado']]
            within_tolerance = [is_within_tolerance(row['premio_ajustado'], p) for p in other_premios]
            
            # If the difference between 'premio_ajustado' and any other value is less than 0.1,
            # check if they are from different 'seguradora' groups, and set 'preco_validado' to 'validado'
            if any(within_tolerance):
                other_seguradoras = [s for s in group_df['seguradora'].unique() if s != row['seguradora']]
                same_seguradora = [s for s in other_seguradoras if str(s).startswith(str(row['seguradora']))]
                if len(same_seguradora) == 0:
                    group_df.at[idx, 'preco_validado'] = 'validado'
                else:
                    group_df.at[idx, 'preco_validado'] = 'duplicado'
            else:
                # Otherwise, set 'preco_validado' to 'nao_validado'
                group_df.at[idx, 'preco_validado'] = 'nao_validado'

    
    # Update the group in the original dataframe with the new 'preco_validado' values
    df_filtered2_filtrados.loc[group_df.index, 'preco_validado'] = group_df['preco_validado']
```
